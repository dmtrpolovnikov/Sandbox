```python
import yfinance as yf
import pandas as pd
```

```python
intrvl = "1d"
data = '2021-01-01'
data1 = '2021-02-02'
price = []

tickers = ['MSFT','BRK-A','TSM','UNH','DIS','BAC','HD','PG','VZ','XOM','KO',
           'T', 'V', 'JPM','VTI','JNJ','WMT','SPY','MA','BML-PJ']

for tics in tickers:
    data = pd.DataFrame(yf.download(tickers=tics, start=data, end=data1, interval=intrvl))
    #price.append([tics, data['Adj Close'].iloc[0]])
    print(data)
```

```python
prc = pd.DataFrame(price)
prc
```
